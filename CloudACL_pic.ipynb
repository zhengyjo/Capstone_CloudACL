```python
df1 = pd.read_csv('/Users/zhengyjo/Desktop/DS501/CAPSTONE/China/ipAgg1_3_6_214.csv')
```

```python
df2 = pd.read_csv('/Users/zhengyjo/Desktop/DS501/CAPSTONE/China/ipAgg1_3_6_216.csv')
```

```python
df3 = pd.read_csv('/Users/zhengyjo/Desktop/DS501/CAPSTONE/China/ipAgg_China_0306P3.csv')
```

```python
df_new = pd.concat([df1, df2,df3])
df_new
```

```python
df_new = df_new.fillna(0)
```

```python
df_new
```

```python
df_new.index.name = 'ipnum'
```

```python
indexed_df = df_new.set_index(['ipnum'])
```

```python
indexed_df
```

```python
df_new_group = indexed_df.groupby(level=['ipnum']).sum()
```

```python
df_new_group
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans()

kmeans.fit(df_new_group)
```

```python
print "cluster centers:"
print kmeans.cluster_centers_
```

```python
features = list(df_new_group.columns.values) 
top_centroids = kmeans.cluster_centers_.argsort()[:,-1:-11:-1] 
print "top features for each cluster:" 
for num, centroid in enumerate(top_centroids): 
    print "%d: %s" % (num, ", ".join(features[i] for i in centroid))
```

```python
df_copy = df_new_group.copy()
```

```python
del df_copy['cat_unknown']
```

```python
kmeans = KMeans(4)
kmeans.fit(df_copy)
```

```python
features = list(df_copy.columns.values) 
top_centroids = kmeans.cluster_centers_.argsort()[:,-1:-10:-1] 
print "top features for each cluster:" 
for num, centroid in enumerate(top_centroids): 
    print "%d: %s" % (num, ", ".join(features[i] for i in centroid))
```
